- Qué variables cualitativas tienen muchos valores únicos (alta cardinalidad)?

primero buscamos cuales con mis variables cualitativas, por ahora suponemos que las variables cualitativas son aquellas que su tipo de datos es una cadena de texto, pero como sabemos que la columna "Lat Long" es la combinación de las columnas "Latitude" y "Longitude" la vamos a eliminar para efectos practicos en un nuevo dataframe y asi mismo vamos a hacer con las siguientes columnas:

- CustomerID

```python
new_df = df.drop(columns=["Lat Long"])
new_df = new_df.drop(columns=["CustomerID"])
```

```python
select_cols = new_df.columns[
    new_df.dtypes.isin(["object", "string"]) & (new_df.nunique() > 2)
]
```

```python
select_cols
```

```python
df_filtered = df[select_cols]
```

```python
df_filtered.nunique()
```

Lo que acabamos de hacer es utilizar operaciones lógicas para obtener las columans del tipo de datos string que tienen más de dos datos únicos, y nos da como resultado que City, Churn Reason son las columnas con mayor cardinalidad (siendo City la de mayor cardinalidad)
